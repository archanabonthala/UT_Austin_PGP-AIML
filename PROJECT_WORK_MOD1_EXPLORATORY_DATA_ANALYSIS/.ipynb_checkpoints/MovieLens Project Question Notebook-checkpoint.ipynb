### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.

```python
#your code here

genre_columns =['unknown', 'Action',
       'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary',
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'] #created list of genres column names to facilitate interation

                  
movie_groupby = movie.groupby('release_year')[genre_columns].apply(sum,axis=0) # sum of movies per genre per year
movie_groupby['total'] = movie_groupby[genre_columns].apply(sum,axis=1)# total number of movie in one year

genre_percent = (movie_groupby[genre_columns].div(movie_groupby['total'], axis= 0) * 100)  #calculating percentage for each genre
genre_percent['release_year'] = genre_percent.index


for col in genre_columns:
    plt.figure(figsize=(15,5))  # set fig size
    plt.xticks(rotation = 60)   # rotate x axis labels to 60 deg
    g = sns.pointplot(x=genre_percent.index,y=col,data=genre_percent)
    g.set_ylim(0, 100)   # set all graphs with same y index to facilitate comparition.
  
```

**Insights:**

Insights on some of the major genres having more number of movie counts
1. Drame: <br>
    i) Looks like this genre appears to have popularity in early years between 1992 -1931. <br>
    ii) After that, this genre show decent popularity with few lows in between.  <br>
    iii) But in recent years, the popularity numbers are incresing with no major down fall.<br>
    
2. Action:<br>
    i) This genre shows least popularity in the early years (1992- 1936).<br>
    ii) There was a suggen spike in popularity between years(1937-1939).<br>
    iii) This genre popularity shows increase from 1976. (although there is fall again in the year 1985)<br>
    
3. Comedy <br>
    i) After drama genre in line, Comedy genre shows huge popularity.<br>
    ii) Though there were lows in between years, this genre shows increase in percentage in between.
    




### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#your code here

rating_copy = rating.copy()    # work on copy of pandas dataframe.
rating_copy['rating_avg'] = rating_copy['rating']
rating_copy['rating_count'] = rating_copy['rating']
rating_summary = rating_copy.groupby(by='movie id').agg({'rating_avg':'mean', 'rating_count':'count'}) # find mean and count of ratings
rating_count_gre_100 = rating_summary[rating_summary['rating_count']>100]  #slice movies which have ratings more than 100

rating_count_avg_sort = rating_count_gre_100.sort_values(by=['rating_avg'], ascending = False) # sort rating average in descending order
rating_output = rating_count_avg_sort.head(25).reset_index() #get top 25 movie ids

final_output = pd.merge(rating_output,movie[['movie id','movie title']], how='left', on = 'movie id') # merge with movie data to get movie titles
final_output
```

### 10. Check for the validity of the below statements with respect to the data provided - 10 marks

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women

**compare the percentages**

**Please pay attention to what should be the denominator while calculating percentages**

1. Merge all the datasets

2. There is no need to conduct statistical tests around this. Just **compare the percentages** and comment on the validity of the above statements.

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
#your code here

#print(rating.columns,user.columns,movie.columns)

# merging data


merge1 = pd.merge(rating,user,how='left',on='user id')  # rating-user left join
merge2 = pd.merge(merge1,movie,how='left',on='movie id') # merge1(rating-user left join)-movie left join

merge2_groupby = merge2.groupby('gender')[genre_columns].apply(sum,axis=0)  # sum of ratings per gender per genre
merge2_groupby.loc['F','total_ratings'] = merge2['gender'].value_counts()['F'] # total number of ratings by Female
merge2_groupby.loc['M','total_ratings'] = merge2['gender'].value_counts()['M'] # total number of ratings by Male

merge2_gender_pert = merge2_groupby.copy() # work on copy of pandas dataframe.
merge2_gender_pert = merge2_gender_pert[genre_columns].div(merge2_gender_pert['total_ratings'],axis =0)*100  # find percentage of each genre for male and female

print(merge2_gender_pert.T)


```

```python
#Other approach : Results are same.


merge1 = pd.merge(rating,user,how='left',on='user id')
merge2 = pd.merge(merge1,movie,how='left',on='movie id')
# Validation of merge , one id is missing so few columns have NaN value
merge2.isna().sum()


total_women_ratings = merge2['gender'].value_counts()['F'] # number of ratings by women
total_men_ratings = merge2['gender'].value_counts()['M']  # number of ratings by men
print('Total ratings received by women :',total_women_ratings)
print('Total ratings received by men :',total_men_ratings)
print('\n\n')


#1. Men watch more drama than women

women_for_drama = merge2.groupby('gender').sum()['Drama']['F'] # number of ratings by women for Drama
men_for_drama = merge2.groupby('gender').sum()['Drama']['M']  # number of ratings by men for Drama
print('Number of drama ratings by women',women_for_drama)
print('Number of drama ratings by men',men_for_drama)

women_drame_percent = (women_for_drama/total_women_ratings)*100  # find 
men_drame_percent = (men_for_drama/total_men_ratings)*100
print('women_drame_percent',women_drame_percent)  # women drama percent
print('men_drame_percent',men_drame_percent)      # men drama percent
print('\n')


#2. Women watch more Sci-Fi than men

women_for_SciFi = merge2.groupby('gender').sum()['Sci-Fi']['F']   # number of ratings by women for Sci-Fi
men_for_SciFi = merge2.groupby('gender').sum()['Sci-Fi']['M']      # number of ratings by men for Sci-Fi
print('Number of Sci-Fi ratings by women',women_for_SciFi)
print('Number of Sci-Fi ratings by men',men_for_SciFi)

women_SciFi_percent = (women_for_SciFi/total_women_ratings)*100
men_SciFi_percent = (men_for_SciFi/total_men_ratings)*100
print('women_SciFi_percent',women_SciFi_percent)     # women SciFi percent
print('men_SciFi_percent',men_SciFi_percent)      # men SciFi percent
print('\n')


#3. Men watch more Romance than women

women_for_romance = merge2.groupby('gender').sum()['Romance']['F']  # number of ratings by women for Romance
men_for_romance = merge2.groupby('gender').sum()['Romance']['M']    ## number of ratings by men for Romance
print('Number of Romance ratings by women',women_for_romance)
print('Number of Romance ratings by men',men_for_romance)

women_romance_percent = (women_for_romance/total_women_ratings)*100  # women romance percent
men_romance_percent = (men_for_romance/total_men_ratings)*100     # men romance percent
print('women_romance_percent',women_romance_percent)
print('men_romance_percent',men_romance_percent)
print('\n')
```
