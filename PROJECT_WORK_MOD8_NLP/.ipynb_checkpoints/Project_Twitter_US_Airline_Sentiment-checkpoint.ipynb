### 4. Text pre-processing: Data preparation. (12 Marks)
NOTE:- Each text pre-processing step should be mentioned in the notebook separately.<br>
     a. Html tag removal.<br>
     b. Tokenization.<br>
     c. Remove the numbers.<br>
     d. Removal of Special Characters and Punctuations.<br>
     e. Removal of stopwords<br>
     f. Conversion to lowercase.<br>
     g. Lemmatize or stemming.<br>
     h. Join the words in the list to convert back to text string in the data frame. (So that each row
          contains the data in text format.)<br>
     i. Print the first 5 rows of data after pre-processing.

```python
## a. Html tag removal.
pd.set_option('display.max_colwidth', 100)

def strip_html_tags(text):
    soup = BeautifulSoup(text, "html.parser")   # define BeautifulSoup object
    stripped_text = soup.get_text()   # remove html tags
    return stripped_text
data['text'] = data['text'].apply(strip_html_tags)   # apply strip_html_tags function to pandas column

print('After processing')
data.head()
```

```python
## c. Remove the numbers.

def remove_digits(text):
    pattern = r'[0-9]'
    text = re.sub(pattern, '', str(text))  # remove digits using regular expression patterns.
    return text

data['text'] = data['text'].apply(remove_digits)  # apply remove_digits function to pandas column
print('After processing')
data.head()
```

```python
#remove special characters

def remove_special_characters(text):
    pattern = r'[^a-zA-z0-9\s]'
    text = re.sub(pattern, '', text)   # remove special characters using regular expression patterns.
    return text

data['text'] = data['text'].apply(remove_special_characters)  # apply remove_special_characters function to pandas column

print('After processing')
data.head()
```

```python
## b. Tokenization.

def Tokenize_text(text):
    tokenizer=ToktokTokenizer()   # break the data into tokens
    tokens=tokenizer.tokenize(text)
    return tokens

data['text'] = data['text'].apply(Tokenize_text) # apply Tokenize_text function to pandas column


print('After processing')
data.head()
```

```python
## to lower

def to_lower(text):
    mod_text = []
    for word in text:
        lower_word = word.lower()   # convert all tokens to lower letters
        mod_text.append(lower_word)
    return mod_text

data['text'] = data['text'].apply(to_lower)  # apply to_lower function to pandas column


print('After processing')
data.head()
```

```python
# remove stop words
def remove_stop_words(text):       
    stopword_list = nltk.corpus.stopwords.words('english')
    filtered_data=[token for token in text if token not in stopword_list]
    return filtered_data

data['text'] = data['text'].apply(remove_stop_words)

print('After processing')
data.head()
```

```python
# g. Lemmatize or stemming.
# h. Join the words in the list to convert back to text string in the data frame. (So that each row contains the data in text format.

nlp = spacy.load('en_core_web_sm')
def lemmatize_text(text_list):   # function to lemmatize text
    text = ' '.join(text_list)
    text = nlp(text)
    text = ' '.join([word.lemma_ if word.lemma_ != '-PRON-' else word.text for word in text])
    return text

data['text'] = data['text'].apply(lemmatize_text)  # apply lemmatize_text function on pandas df column

print('After processing')
data.head()
```

### Modelling , tuning and Evaluation
- Fit the model using vectorized column - Tune the model to improve the accuracy - Evaluate the model using the confusion matrix - Print the top 40 features and plot their word cloud using both types of vectorization. (7+7 Marks)

```python
##  Function to calculate different metric scores of the model - Accuracy, Recall and Precision
def calc_metrics_score(model,X_train,y_train,X_test,y_test,flag=True):
    '''
    model : model classifier to predict values of X

    '''
    # defining an empty list to store train and test results
    score_list=[] 
    
    #Predicting on train and tests
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)
    
    #Accuracy of the model
    train_acc = model.score(X_train,y_train)
    test_acc = model.score(X_test,y_test)
    
    
    score_list.extend((train_acc,test_acc))
        
    # If the flag is set to True then only the following print statements will be dispayed. The default value is set to True.
    if flag == True: 
        print("Accuracy on training set : ",model.score(X_train,y_train))
        print("Accuracy on test set : ",model.score(X_test,y_test))

    
    return score_list # returning the list with train and test scores
```

```python
## Function to draw confusion matrix
def draw_confusion_matrix(model,X_test,y_actual,labels=[-1,0,1]):
    '''
    model : model classifier
    y_actual : ground truth  
    
    '''
    y_predict = model.predict(X_test)  # get predictions for given test data
    cm=confusion_matrix( y_actual, y_predict, labels=[-1, 0, 1])   # call confusion matrix function
    df_cm = pd.DataFrame(cm, index = [i for i in ["Actual - Negative","Actual - Nuetral","Actual - Positive"]],
                  columns = [i for i in ['Predicted - Negative','Predicted - Nuetral','Predicted - Positive']])    # define pandas df with required index and columns
    group_counts = ["{0:0.0f}".format(value) for value in
                cm.flatten()]
    group_percentages = ["{0:.2%}".format(value) for value in
                         cm.flatten()/np.sum(cm)]       # calculate TP,FP,TN,FN percentages
    labels = [f"{v1}\n{v2}" for v1, v2 in
              zip(group_counts,group_percentages)]
    labels = np.asarray(labels).reshape(3,3)
    plt.figure(figsize = (10,7))   # set figure size
    sns.heatmap(df_cm, annot=labels,fmt='')   # draw heat map
    plt.ylabel('True label')       # set y label
    plt.xlabel('Predicted label')  # set x label
```
