```python
import pandas as pd
import numpy as np
import random
```

```python
seq = [1,2,3,4,5,6]
list_out =[]

output = [] 
def reemovNestings(l): 
    for i in l: 
        if type(i) == list: 
            reemovNestings(i) 
        else: 
            output.append(i)
    return output



list_row = []
df = pd.DataFrame( columns = ['Sample','Throw1','Throw2','Throw3','Throw4'])
for i in range(1,31,1):
    l =random.choices(seq, k =4)
    df.loc[i]=[i]+l


   
#df = pd.DataFrame(df, columns = ['Sample','Throw1','Throw2','Throw3','Throw4'])
df['Mean'] = df[['Throw1','Throw2','Throw3','Throw4']].mean(axis =1)
```
