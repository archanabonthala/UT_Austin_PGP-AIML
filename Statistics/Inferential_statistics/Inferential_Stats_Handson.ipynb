```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Binoial distribution
n=7
p=.6
```

```python
k = np.arange(0,8,1)
k
```

```python
binomial = stats.binom.pmf(k,n,p)
```

```python
binomial
```

```python
plt.plot(k,binomial,'o-')
plt.title('Binomial: n=%i, p=%.2f' %(n,p), fontsize=15)
plt.xlabel('Number of Successes')
plt.ylabel('Probability of Successes')
plt.show()
```

```python
sum(binomial)
```

```python
#poisson distribution
rate = 3
n = np.arange(0,20)
```

```python
poisson = stats.poisson.pmf(n,rate)
```

```python
poisson
```

```python
poisson[4]
```

```python
# probability of getting more than three customers
1-(poisson[0]+poisson[1]+poisson[2]+poisson[3])
```

```python
plt.plot(n, poisson,'o-')
plt.title('Poisson: $\lambda$ = %i' %rate)
plt.xlabel('Number of customers arriving in a minute')
plt.ylabel('Probability of Number of customers arriving in a minute')
plt.show()
```

```python
#Normal Distribution

mean = 0.295
sd = 0.025

z= (0.280-mean)/sd
```

```python
stats.norm.cdf(z)
```

```python
stats.norm.cdf(0.280,loc = 0.295, scale=0.025)
```

```python
# area to the right ( more than 0.280)

1- stats.norm.cdf(0.280,loc = 0.295, scale=0.025)
```

```python
1- stats.norm.cdf(0.350,loc = 0.295, scale=0.025)
```

```python
# chance of between 260 gms and 340 gms
stats.norm.cdf(0.340,loc = 0.295, scale=0.025) - stats.norm.cdf(0.260,loc = 0.295, scale=0.025)
```
